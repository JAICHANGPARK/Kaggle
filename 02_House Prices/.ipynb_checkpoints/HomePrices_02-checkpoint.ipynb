```python
labels = []
values = []
```

## id col remove

```python
num_feat=train.columns[train.dtypes!=object]
print(num_feat)
num_feat=num_feat[1:-1] 
print(num_feat)
```

```python
for col in num_feat:
    labels.append(col)
    values.append(np.corrcoef(train[col].values, train.SalePrice.values)[0,1])
print(labels)
print(values)
```

```python
# lable의 개수 
leng = len(labels)
print(leng)
# arrange로 정렬 
ind = np.arange(len(labels))
print(ind)
```

```python
correlations=train.corr()
attrs = correlations.iloc[:-1,:-1] # all except target
print(correlations)
print(attrs)
```

```python
threshold = 0.5
```

```python
important_corrs = (attrs[abs(attrs) > threshold][attrs != 1.0]).unstack().dropna().to_dict()
print(important_corrs)
```

```python
unique_important_corrs = pd.DataFrame(
    list(set([(tuple(sorted(key)), important_corrs[key]) \
    for key in important_corrs])), 
        columns=['Attribute Pair', 'Correlation'])
```

- Multicollinearity occurs when your model includes multiple factors that are correlated not just to your target variable, but also to each other.

```python
# sorted by absolute value
unique_important_corrs = unique_important_corrs.ix[
    abs(unique_important_corrs['Correlation']).argsort()[::-1]]

unique_important_corrs
```

heatmap

```python
corrMatrix=train[["SalePrice","OverallQual","GrLivArea","GarageCars",
                  "GarageArea","GarageYrBlt","TotalBsmtSF","1stFlrSF","FullBath",
                  "TotRmsAbvGrd","YearBuilt","YearRemodAdd"]].corr()

sns.set(font_scale=1.10)
plt.figure(figsize=(10, 10))

sns.heatmap(corrMatrix, vmax=.8, linewidths=0.01,
            square=True,annot=True,cmap='viridis',linecolor="white")
plt.title('Correlation between features');
```
